```python
# Look at all the numbers which are positive and than find it's even or odd.
# In case of negative I will ignore that number

numbers_list = [10, -4, 30, 27, 35, -33, 45, 48]


for number in numbers_list:
    print(f'Looking at ........ {number}')
    if number < 0:
        continue
    remainder = number%2
    if remainder == 0:
        print('Even Number')
    else:
        print('Odd Number')
```

```python
def find_percentage(list_of_marks):
    # assume every subject is of 100 marks
    obtained_marks = 0
    for marks in list_of_marks:
        obtained_marks = obtained_marks + marks
    no_of_subjects = len(list_of_marks)
    total_marks = no_of_subjects*100
    percentage = (obtained_marks/total_marks)*100
    return percentage
```

```python
marks_of_arun = [100, 39, 98, 59]

percentage_of_arun = find_percentage(marks_of_arun)

print(percentage_of_arun)
```

```python
marks_of_prasenjit = [90, 43, 92, 67, 89]

percentage_of_prasenjit = find_percentage(marks_of_prasenjit)

print(percentage_of_prasenjit)
```

```python
marks_of_shweta = [90, 43, 92, 67, 89, 100]

percentage_of_shweta = find_percentage(marks_of_shweta)

print(percentage_of_shweta)
```

```python
def find_largest_number(num1, num2, num3):
    largest_so_far = None
    if num1 > num2:
        largest_so_far = num1
    else:
        largest_so_far = num2
    if largest_so_far < num3:
        largest_so_far = num3
    return largest_so_far
```
