```python
def happy(num, x=None):
    if x is None:
        x = []
    for i in range(num):
        x.append(i)
    print(x)
```

```python
# Dictionary & Sets
# hashing

x = ['mohan', 'rohan', 'amit', 'niranjan']

# find if 'amit' is present

for student in x:
    if student == 'amit':
        print('found amit')
```

```python
# key, value pairs

x = {"mohan" : True, "rohan": False, "amit": True, "niranjan": False}

# find if amit is present

return_value = x.get('amit')

print(return_value)
```

```python
big_list = []
big_dict = {}

for num in range(100000000):
    big_list.append(num)
    big_dict[num] = True
```

```python
len(big_list)
```

```python
len(big_dict)
```

```python
search_element = 99999995

import time

start_time = time.time()

for num in big_list:
    if num == search_element:
        print('found')
        break

end_time = time.time()

time_diff = end_time-start_time

print(f"{time_diff} seconds")
```

```python
search_element = 99999995

import time

start_time = time.time()

if big_dict.get(search_element):
    print('found')

end_time = time.time()

time_diff = end_time-start_time

print(f"{time_diff} seconds")
```
