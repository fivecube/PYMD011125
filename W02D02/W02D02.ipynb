```python
# Scope and Namespaces

x = 10

def fun1():
    # local namespace
    x = 2
    print(f'x is {x}')

fun1()

print(f'The value of x outside is {x}')
```

```python
# global keyword
x = 10

def fun1(num):
    global x
    x = num
    return num

fun1(4)

fun1(100)
```

```python
# OBJECT ORIENTED PROGRAMMING

# CLASS IS LIKE A BLUE PRINT OF A HOUSE


class StudentMarks:
    def __init__(self, name, maths, english):
        print('INIT FUN IS CALLED', name, id(self))
        self.name = name
        self.maths = maths
        self.english = english

    def find_total(self):
        return self.maths + self.english
```

```python
# This line creates an object of StudentMarks class
# object is an actual house created from the blue print

john_marks = StudentMarks(name="john", english=400, maths=300)
print(id(john_marks))
```

```python
# Encapsulation

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance
```

```python
acc = BankAccount(1000)
```

```python
acc.__deposit(100)
```

```python
acc.get_balance()
```

```python
acc.withdraw(200)
```

```python
acc.get_balance()
```

```python
acc.get_balance()
```

```python
class Coordinate:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, another_coordinate):
        return Coordinate(self.x + another_coordinate.x, self.y + another_coordinate.y)

    def __sub__(self, co2):
        return Coordinate(self.x - co2.x, self.y - co2.y)
```

```python
co1 = Coordinate(2,3)
```

```python
co2 = Coordinate(10,20)
```

```python
co3 = co1 + co2
```

```python
co3 = co1 - co2
```

```python
print(co3.x, co3.y)
```
