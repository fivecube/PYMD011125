```python
# Logical and, or, not


isMathsPassed = True
isEnglishPassed = True
isSocialSciencePassed = False

# If everything is True, "and" gives us True
# If any one of the thing is False, "and" will give us False


mohit_passed = isMathsPassed and isEnglishPassed and isSocialSciencePassed

print(mohit_passed)
```
