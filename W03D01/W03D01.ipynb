```python
class Vehicle:
    all_cars = [] # class attributes : common for all the objects.

    def __init__(self, color, power):
        self.color = color
        self.power = power
        self.speed = 0

    @staticmethod
    def new_happy_fun(x, y):
        return x + y

    @classmethod
    def keep_entry_of_all_cars(cls, car_object):
        cls.all_cars.append(car_object)

    def get_color(self):
        return self.color

    def set_color(self, color):
        self.color = color

    def apply_brakes(self):
        self.speed = 0
```

```python
class Car(Vehicle):
    def apply_brakes(self):
        self.speed -= 50
```

```python
toyota = Car('Black', '2')
```

```python
class Bike(Vehicle):
    def __init__(self, name, color, power):
        super().__init__(color, power)
        self.name = name
```

```python
# csv

import csv

# comma separated values

# name, roll_no, class
# rohan, 45, 10
# mohan, 35, 9

rows = [['name', 'age'], ['rohan', 40], ['mohan', 50]]

with open('people.csv', 'w') as fp:
    writer = csv.writer(fp)
    writer.writerows(rows)
```

```python
class WalletBalanceLow(Exception):
    pass
```

```python
with open('wallet_balance.txt', 'w') as fp:
    fp.write('40')
```

```python
def give_my_balance():
    with open('wallet_balance.txt', 'r') as fp:
        content = fp.read()
        wallet_balance = int(content)
    return wallet_balance

try:
    x = int(input('How many ice creams you want!'))
    amount = 10*x
    wallet_balance = give_my_balance()
    if wallet_balance < amount:
        raise WalletBalanceLow
except WalletBalanceLow as exc:
    print('Your balance is not sufficient to order than many ice creams')
```

```python
class AutoCloseOpen:
    def __enter__(self, name_of_file, mode_of_operation):
        self.fp = open(name_of_file, mode_of_operation)
        return self.fp

    def __exit__(self, exc_type, exc, tb):
        self.fp.close()
```
